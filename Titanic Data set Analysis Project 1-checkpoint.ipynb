# total passengers survived in numbers

```python
dd_sur= df[['Survived', 'PassengerId']]   
dds= dd_sur.groupby(['Survived'])
```

```python
total_survived_in_Numbers = dds.count()
total_survived_in_Numbers
```

# total passengers Killed percentage.

```python
total_dead_Percentage = (total_survived_in_Numbers['PassengerId'][0]/total_survived_in_Numbers['PassengerId'].sum())*100
print('Total',total_dead_Percentage.round(), '% were dead in Titanic.')
```

# total passengers survived percentage.

```python
total_survived_Percentage = (total_survived_in_Numbers['PassengerId'][1]/total_survived_in_Numbers['PassengerId'].sum())*100
print('Total',total_survived_Percentage.round(), '% were Survived in Titanic.')
```

# 4) Which age group, survived the most, kids (0-12), teen agers (13-19), youngsters (20-30), middle age people (31-50) or elders (50 and)  greater) ?

```python
dd_sur_age= dd[['Survived', 'Age']]   
dd_sur_age= dd_sur_age.groupby(['Age'],as_index = False)
```

```python
dd_sur_age = dd_sur_age.count()
dd_sur_age
```

# Kids survived in titanic.

```python
dd_sur_kids = dd_sur_age[dd_sur_age["Age"].between(0, 12)]
dd_sur_kids = dd_sur_kids['Age'].count()
print('Total ',dd_sur_kids, 'Kids with age group Of ( 0 to 12 ) were survived.')
```

# teenagers survived in titanic.

```python
dd_sur_teens = dd_sur_age[dd_sur_age["Age"].between(13, 19)]
dd_sur_teens = dd_sur_teens['Age'].count()
print('Total ',dd_sur_teens, 'teen agers with age group of ( 13 to 19 ) were survived.')
```

# youngesters survived in titanic.

```python
dd_sur_youngsters = dd_sur_age[dd_sur_age["Age"].between(20, 30)]
dd_sur_youngsters = dd_sur_youngsters['Age'].count()
print('Total ',dd_sur_youngsters, 'youngsters with age group of ( 20 to 30 ) were survived.')
```

# Middle age passengers survived  in titanic.

```python
dd_sur_middle = dd_sur_age[dd_sur_age["Age"].between(31, 50)]
dd_sur_middle = dd_sur_middle['Age'].count()
print('Total ',dd_sur_middle, 'middle age with age group of ( 31 to 50 ) were survived.')
```
